Train-Test Split

```python
# === Create features X and target y; preprocess and split (70/30, random_state=116) ===
# Use 'status' as target (acquired=1, closed=0)
if 'status' not in data.columns:
    raise KeyError("Column 'status' not found in `data`  pick a valid target column")

y = data['status'].map({'acquired': 1, 'closed': 0})
if y.isnull().any():
    # fallback to factorize if there are unexpected labels
    print("Warning: some status values were not in the mapping  falling back to numeric factorization")
    y = pd.factorize(data['status'])[0]

# build X by dropping identifiers/leaky columns
drop_cols = ['id','name','status','Unnamed: 0','state_code.1','founded_at','closed_at','category_code','object_id']
drop_cols = [c for c in drop_cols if c in data.columns]
X = data.drop(columns=drop_cols).copy()

# one-hot encode low-cardinality categoricals
cat_cols = [c for c in ['State','category'] if c in X.columns]
if cat_cols:
    X = pd.get_dummies(X, columns=cat_cols, dummy_na=False)

# normalize boolean-like columns to 0/1
bool_cols = ['has_VC','has_angel','has_roundA','has_roundB','has_roundC','has_roundD','is_CA','is_NY','is_MA','is_TX','is_otherstate']
for c in bool_cols:
    if c in X.columns:
        X[c] = pd.to_numeric(X[c], errors='coerce').fillna(0).astype(int)

# impute numeric columns with median
num_cols = X.select_dtypes(include=['number']).columns
X[num_cols] = X[num_cols].fillna(X[num_cols].median())

# drop remaining non-numeric high-cardinality columns
non_num = X.select_dtypes(exclude=['number']).columns.tolist()
for c in non_num:
    if X[c].nunique() <= 50:
        X = pd.get_dummies(X, columns=[c], dummy_na=False)
    else:
        X = X.drop(columns=[c])

# split
from sklearn.model_selection import train_test_split
_stratify = y if (pd.Series(y).nunique() < 20) else None
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=116, stratify=_stratify)

print('X shape:', X.shape)
print('x_train shape:', x_train.shape, 'x_test shape:', x_test.shape)
print('y distribution:')
print(pd.Series(y).value_counts(normalize=True))
```

Model Building

Define Preprocessing and Model Function

```python
rf = RandomForestClassifier()
param_grid = {'n_estimators': [100,200,300],'max_depth': [10,20,30],
'min_samples_split': [2,4,6],'min_samples_leaf': [1,2,3],
'bootstrap': [True, False]}
grid_search = GridSearchCV(estimator=rf, param_grid = param_grid, cv = 5, n_jobs = 1, verbose = False)
grid_search.fit(x_train, y_train)
print('Best parameters:', grid_search.best_params_)
```

```python
rf = RandomForestClassifier(n_estimators=100, bootstrap=True, max_depth=10,min_samples_leaf=1, min_samples_split=2)
model_rf = rf.fit(x_train,y_train)
y_pred_rf = model_rf.predict(x_test)
cr_rf= classification_report(y_pred_rf, y_test)
print(cr_rf)
```

Performance Testing

Testing Accuracy

```python
#applying Random forest classifier
model=RandomForestClassifier()
model.fit(x_train, y_train)
y_pred_test=model.predict(x_test)
y_pred_train=model.predict(x_train)
```

```python
#checking accuracy
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
test_acc = accuracy_score(y_test,y_pred_test)
train_acc = accuracy_score(y_train,y_pred_train)
print('test_acc:', test_acc)
print('train_acc:', train_acc)
```

The Random Forest classifier achieved an accuracy of 100% on the test dataset (test_acc). This means that the model correctly predicted the target variable for 100% of the test instances. However, the model achieved a perfect accuracy of 100% on the training dataset (train_acc), indicating that it may be overfitting the training data. Overfitting occurs when a model performs exceptionally well on the training data but fails to generalize well to unseen data. Further evaluation and tuning may be necessary to address the potential overfitting issue.
